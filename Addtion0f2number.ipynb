```python
import pandas as pd
```

```python
data=pd.read_csv('add.csv')
```

```python
data.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['x'],data['sum'])
```

```python
x=data[['x','y']]
y=data['sum']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
df=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

```python
from tkinter import *
import joblib
```

```python
def show_entry_fields():
    
    p1=float(e1.get())
    p2=float(e2.get())

    reg=joblib.load('Adding_two_Number')
    result=reg.predict([[p1,p2]])

    Label(master,text='Sum is =').grid(row=4)
    Label(master,text=result).grid(row=5)
                                  
    print("Sum is:",result)


master=Tk()
master.title("Addition of two number") 

label=Label(master,text='Addition of two number uisng ML',bg='white',fg='blue').grid(row=0,columnspan=2)

Label(master,text='Enter the First Number:').grid(row=1)
Label(master,text='Enter the Second Number:').grid(row=2)

e1=Entry(master)
e2=Entry(master)

e1.grid(row=1,column=1)
e2.grid(row=2,column=1)

Button(master,text='Predict',command=show_entry_fields).grid(row=3,columnspan=2)

mainloop()        
```
